```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data = pd.read_csv('auto_insurance_sweden.csv') #cargar archivo csv
x = data['reclamos'] #asignamos x a la columna de reclamos
y = data['SK'] #asignamos y a la columna de coronas
```

```python
plt.scatter(x,y)
plt.xlabel('reclamos')
plt.ylabel('costo en SK')
plt.show()
```

```python
def reg_lineal(x,y):
  #calcular medias
  x_mean = x.mean()
  y_mean = y.mean()

  theta_1_num = ((x - x_mean) * (y - y_mean)).sum()
  theta_1_den = ((x - x_mean)**2).sum()
  theta_1 = theta_1_num / theta_1_den

  theta_0 = y_mean - (theta_1 * x_mean)

  regression_line = 'y = {} + {}x'.format(theta_0, round(theta_1, 3))

  return(theta_0, theta_1, regression_line)
```

```python
theta_0, theta_1, reg_line = reg_lineal(x,y)
print('Regresion Lineal: ', reg_line)
```

```python
plt.figure(figsize=(12,5))
plt.scatter(x, y, s = 300, linewidths=1, edgecolors='black')
plt.title('Medir costo de reclamos de accidentes automovilisticos')
plt.xlabel('Numero de Reclamos', fontsize = 15)
plt.ylabel('Costo en Coronas Suecas', fontsize = 15)
plt.plot(x, theta_0 + theta_1*x, c ='r', linewidth =5, alpha = .5, solid_capstyle='round')
```

```python
# y = theta_0 + theta1 * x
def prediccion (theta_0, theta1, nuevo_x):
  cost = theta_0 + theta_1*nuevo_x
  return(cost)  

prediccion(theta_0, theta_1, 115)
```
